# Frequency Distribution Table

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'Product':['Box','Bottles','Pen','Markers'],
        'Sales':[14,24,31,12]}
```

```python
df1=pd.DataFrame(data, columns=['Product','Sales'])
print(df1)
ax = df1.plot.bar(x='Product', y='Sales', rot=0)
```

```python
ax = df1['Sales'].plot.pie(x='Product', figsize=(5, 5))
```

```python
from matplotlib.ticker import PercentFormatter

df = pd.DataFrame({'country': [177.0, 7.0, 4.0, 122.0, 2.0, 191.0]})
df.index = ['USA', 'Canada', 'Russia', 'UK', 'Belgium', 'Mexico']
df = df.sort_values(by='country',ascending=False)
df["cumpercentage"] = df["country"].cumsum()/df["country"].sum()*100

fig, ax = plt.subplots()
ax.bar(df.index, df["country"], color="C0")
ax2 = ax.twinx()
ax2.plot(df.index, df["cumpercentage"], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
plt.show()
```
